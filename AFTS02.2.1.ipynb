### 2.1 Stationarity

stationarity = foundation of TSA

**strictly stationary** time series
- if 
- the joint distribution of ($r_{t_1}, ..., r_{t_k}$)
- is identical to 
- the joint distribution of ($r_{t_{1+t}}, ..., r_{t_{k+t}}$)
- for all t
- where k is an arbitrary positive integer
- where (t<sub>1</sub>, ... t<sub>k</sub>) is a collection of k positive integers

so, the joint distribution (r<sub>t<sub>1</sub></sub>, ..., r<sub>t<sub>k</sub></sub>)
- is an sample from time series {r<sub>t</sub>}
- is separated in time by any t from ($r_{t_{1+t}}, ..., r_{t_{k+t}}$)
- is invariant under time shift t
    - invariant = properties unchanged

strict stationarity 
- time invariance is strong condition
- hard to verify empirically

**weakly stationary**
- the mean of time series {$r_t$} is time invariant
- covariance between {$r_t$} and {$r_{t+\ell}$} [exists but] is time invariant
- where $\ell$ is an arbitrary integer
- specifically {$r_t$} time series is weakly stationary
    - if $E(r_t) = \mu$ is a constant
    - if $\gamma_{\ell} = \text{covariance}(r_t,r_{t-\ell})$
        - only depends on $\ell$
        - does not depend on t
        - so $\gamma_{\ell=20}$ may not = $\gamma_{\ell=40}$
        - so $\gamma_{60}$ @ t = 0 must = $\gamma_{60}$  @ t = 40
- example:
    - observe T data points
        - {r| t=1, ..., T}
    - weak stationarity
        - T data points fluctuate 
            - with constant variation 
            - around a fixed level
        - allows inference concerning future observations = prediction
- in the condition of weak stationarity, it is implicitly assumed that 
    - the **first two moments of r<sub>t</sub> are finite**
    - the converse is not true always: if first two moments are finite --> weakly stationary.
    - if the time series is normally distributed then weak stationarity is same as strict stationarity
    
- checking for weakly stationarity: divide the data into subsamples and check consistency of the results obtained across subsamples.

[Variance of the time series cannot change (e.g. grow) over time without causing covariance to change.  If at time t=100 variance is x nd at time t=200 variance grows to be 2x then the covariance of that 100 day lag will change when time t = 200.  That's not allowed under weak stationarity.  Series can be weak but not strictly stationary if covariance t vs t+100 changes over time but the joint distribution does not change. So, e.g. is it possible that covariance falls from x to zero even as the joint distribution remains unchanged? This comes up in later chapters.] 

**autocovariance**
- lag-$\ell$ autocovariance of $r_t = Cov(r_t,r_{t-\ell}) = \gamma_{\ell}$ 
    - property 1: $\gamma_0 = Var(r_t)$
    - property 2: $\gamma_{-\ell} = \gamma_{+\ell}$ holds because
        - $\,Cov(r_t, r_{t-(-\ell)} = Cov(r_{t-(-\ell)},r_t) = Cov(r_{t+\ell}, r_t) = Cov(r_{t_1},r_{t_1-\ell})$ where $\,t_1 = t + \ell$  

## 2.2 Correlation and autocorrelation function

**correlation** between two random variables X and Y:

$\large \bar{x} = \sum_{t=1}^T x_t/T$ = sample mean of Y

$\large \bar{y} = \sum_{t=1}^T y_t/T$ = sample mean of Y

$$\large \,\rho_{x,y} = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)Var(Y)}} = \frac{E[(X-\mu_x)(Y-\mu_y)]}{\sqrt{E(X-\mu_x)^2 E(Y-\mu_y)^2}}$$



where 
- $\mu_x$ and $\mu_y$ are mean of X and Y
- assume:
    - variances exist
- $\rho_{x,y}$ measures linear dependence between X and Y
- $-1 ≤ \rho_{x,y} ≤ 1$ 
- $\rho_{x,y} = \rho_{y,x}$
- $\rho_{x,y} = 0$ indicates variables X and Y are uncorreleated.
- if X and Y are random variables, then $\rho_{x,y} =0$ **IFF X are Y are independent.**
- when a **sample $\{(x_t, y_t)\}_{t=1}^T$** is available, the correlation is consistently estimated with the sample counterpart:

$$\large \,\hat{\rho}_{x,y} = \frac{\sum_{t=1}^T(x_t-\bar{x})(y_t-\bar{y})}{\sqrt{\sum_{t=1}^T(x_t-\bar{x})^2 \sum_{t=1}^T(y_t-\bar{y})^2}}$$



where $\bar{x}$ and $\bar{y}$ are defined above.

The concept of correlation is generalized to **autocorrelation** to compute the linear dependence between:

- $r_{t-i}$

$\rho_{\ell}$ 
- **$lag_{\ell}$ autocorrelation** of $r_t$ 
- the correlation coefficient between $r_t$ and $r_{t-\ell}$
- is a function of "$\ell$" only under the weak stationarity assumption [see note below]
    - [the denominator in (2.1) reduces to $\gamma_0$ because of weak stationarity.]
    - "where the property Var($r_t$) = Var($r_{t-\ell}$) for weak stationarity is used"

(2.1)

$$\large \rho_{\ell} = \frac{\text{Cov}(r_t,r_{t-\ell})}{\sqrt{Var(r_t)Var(r_{t-\ell})}} = \frac{Cov(r_t,r_{t-\ell})}{\text{Var}(r_t)} = \frac{\gamma_{\ell}}{\gamma_0}$$



where: 
- $\rho_0 = 1, \rho_{\ell} = \rho_{-\ell}, -1 ≤ \rho_{\ell} ≤ 1$
- weakly stationay series $r_t$ is not serially correlated if and only if $\rho_{\ell}$ = 0 for ALL periods $\ell > 0$.

**sample autocorrelation**
- sample returns denoted $\{r_{i,t}\}_{t=1}^{t=T}$
- $\bar{r} = \sum_{t=1}^{T} r_t/T$ = sample mean
- lag-1 sample autocorrelation of r<sub>t</sub> is

$$\hat{\rho}_1 = \frac{\sum_{t=2}^{T} (r_t - \bar{r}) (r_{t-1} - \bar{r})}{\sum_{t=1}{T} (r_t - \bar{r})^2}$$



where:
- [denominator is var(r<sub>t</sub>)]

$\hat{\rho}_1$ is a consistent estimate of $\rho_1$ under general conditions:
- if {$r_t$} is an independently and identically distributed (iid) sequence
- if $E(r_t^2) < \infty$
- then $\hat{\rho}_1$ is asymptotically normal with mean zero and variance = 1/T
- and this property can be used to **test the null hypothesis $\,H_0: \rho_1 = 0$** versus the alternative $\,H_1: \rho_1 ≠ 0$
    - test statistic = t-ratio = $\frac{\hat{\rho}_1}{\sqrt{\frac{1}{T}}} = \sqrt{T}\hat{\rho}_1$ that follows asymptotically the standard normal distribution.
    - null hypothesis H<sub>0</sub> is rejected if the t-ratio is large in magnitude or if the p-value of the t-ratio is small < 0.05 e.g.
    
in general, the lag-el sample autocorrelation of r<sub>t</sub> is defined similarly:

(2.2)

$$\hat{\rho}_{\ell} = \frac{\sum_{t=\ell+1}^{T} (r_t - \bar{r}) (r_{t-\ell} - \bar{r})}{\sum_{t=1}^{T} (r_t - \bar{r})^2} ,\,\,\,\, 0 ≤ \ell ≤ T-1$$



$\hat{\rho}_{\ell}$ is asymptotically normal
- with:
    - $\mu = 0$
    - $\sigma^2 = 1/T$ 
- for any fixed positive integer "\ell"
- if {$r_t$} is an **iid sequence** with:
    - $E(r_t^2) < \infty$

more generally according to **Bartlett's formula**: 
- if {$r_t$} is a **weakly stationary time series** satisfying:
    - $\,r_t = \mu + \sum_{i=0}^{q} \psi_i a_{t-i}$, where
        - $\psi_0$ = 1   
            [$\,r_t = \mu + 1 \times a_t + \sum_{i=1}^{q} \psi_i a_{t-i}$]
        - {a<sub>j</sub>} is a sequence of iid random variables with mean zero.
- then $\hat{\rho}_{\ell}$ is asymptotically normal with 
    - $\mu = 0$ 
    - $\sigma^2 = \frac{(1+2\sum_{i=1}^{q} \rho_i^2)}{T}$ 
        - for "$\ell$" > q

With a larger standard error, the t-ratio will be smaller, less likely to be sufficiently far from $\rho_{\ell} = 0$ to cause rejection of null of no autocorrelation, $\rho_{\ell} = 0$. ]

### Testing Individual ACF

$\,H_0: \rho_{\ell} = 0$

$\,H_a: \rho_{\ell} ≠ 0$

test statistic: 

t-ratio = $\frac{\hat{\rho}_{\ell}}{\sqrt{(1+2\sum_{i=1}^{\ell-1} \hat{\rho}_i^2)/T}}$



How to test: 
- If {$r_t$} is stationary Gaussian series satisfying $\rho_j$ = 0 for j > $\ell$, 
- then the t-ratio is asymptotically distributed as a standard normal random variable. [meaning that the longer the time series against which the lag-$\ell$ is tested the more the t-ratio resembles a standard normal random variable.] 
- Hence, the decision rule of the test is to 
    - reject $H_0$ if |t-ratio| > $Z_{\frac{\alpha}{2}}$ 
    - where $Z_{\frac{\alpha}{2}}$ = the 100(1- $\frac{\alpha}{2}$)th percentile of the standard normal distribution.
- software packages use 1/T as the asymptotic variance for $\hat{\rho}_{\ell}$ for all $\ell$ ≠ 0.
    - software packages doing this are assuming that the underlying time series is an iid sequence. [all iid are strong stationary.]
    
Bias of estimating ACF:
- in finite samples, $\hat{\rho}_{el}$ is a biased estimator of $\rho_{el}$. the bias is in the order of 1/T which can be substantial when the size T is small.  
    - [this is why we throw out 1/T estimates that are significant]
    - most financial applications, T is large so bias is not serious.
    
### Portmanteau test

Portmanteau test (Box and Pierce, 1970)
- is the test statistic for testing jointly that several [1 to m sequence of] autocorrelations of $r_t$ are zero.
- $\,H_0: \rho_1 = \rho_2 = ... = \rho_m = 0$
- $\,H_a: \rho_i ≠ 0, i\in\{1, ..., m\}$
- assuming {$r_t$} is an iid sequence with moment conditions
- then $Q^*$(m) is asymptotically a chi-squared random variable with m degrees of freedom.

$$\,Q^*(m) = T \sum_{\ell=1}^m \hat{\rho}_{\ell}^2$$



Ljung and Box (1978)
- modify $Q^*$(m) to increase the power of the test in finite samples
- [ the test statistic divides by $T-\ell$ but also multiplies by T+2 which would make it larger and more likely to cause a rejection of null (more power) that there is no autocorrelation. ]

(2.3)

$$\,Q(m) = T (T + 2) \sum_{\ell=1}^m \frac{\hat{\rho}_{\ell}^2}{T - \ell}$$



- Decision rule:
    - reject $H_0$ if Q(m) > $\chi_a^2$
        - where $\chi_a^2$ denotes the 100(1-$\alpha$)th percentile of a chi-squared distribution with m degrees of freedom.
    - reject $H_0$ if p-value is less than or equal to $\alpha$ = the significance level. (Most software will yield a p value of Q(m))


**m**
- [is the number of autocorrelations tested]
- [is not just the count of autocorrelations since the counter t represents time.]
- [is the terminal in a counted reqular sequence beginning with 1]
- choice of m changes the Q(m) statistic's performance
    - Simulation studies suggest m ~ ln(T) as a rule yields better power performance. [Is this rule related to the bias mentioned earlier?**]
        - modify this general rule **if analyzing seasonal time series** when lags at multiples of seasonality are more important.

### Linear Time Series

A time series r<sub>t</sub> is linear if can be written:

(2.4)

$$\,r_t = \mu + \sum_{i=0}^{\infty} \psi_i a_{t-i}$$

    - is the mean of r<sub>t</sub>
- $\psi_i$ 
    - are coefficients that govern the dynamic structure of r<sub>t</sub>

        - $\psi_i$ for i ≠ 0 
    - are called the $\psi$ weights of r<sub>t</sub>
- a<sub>t</sub> 
    - is a sequence of independent and identically distributed (iid) random variables
    - with mean zero
    - and with a well defined distribution
    - i.e. the sample set {a<sub>t</sub>} is a white noise series.
    - denotes the new information at time t of the time series
        - [new because $\mu$ is already known?]
    - is referred to as **innovation** or **shock** at time t
    - is a **continuous** random variable for this book
    
Not all time series are linear; nonlinearity and nonlinear models are are Chapter 4's remit.

$\large (1 - \phi_1 B - \phi_2 B^2) \rho_{\ell} = 0$

$\large 1 - \phi_1 x - \phi_2 x^2 = 0$

solutions to this equation are 

$\large x = \frac{\phi_1 \pm \sqrt{\phi_1^2 + 4\phi_2}}{-2\phi_2}$

$$\large k = \frac{2\pi}{cos^{-1}[\phi_1/(2\sqrt{-\phi_2})]}$$

$\large \phi_1 = 2a$

$\large \phi_2 = -(a^2 + b^2)$

and

$$\large k = \frac{2\pi}{cos^{-1}[a/(\sqrt{a^2 + b^2})]}$$



where $\sqrt{a^2 + b^2}$ is the absolute value of $\,a \pm bi$.

### Stationarity

AR(2) time series stationarity condition is the absolute values of the 2 characteristic rootes $\,(\omega_1, \omega_2)$ are < 1 = the 2 characteristic roots are < 1 in modulus.  Equivalently, the two solutions to the characteristic equation, $\,(x_1, x_2)$ are > 1 in modulus.  



Under these conditions the recursive (2.13) ensures that ACF of model converges to zero as lag-el increases.  This convergence property is necessary condition for stationary time series.  The condition also applies to AR(1) models where the polynomial equation is $\,1 - \phi_1 x = 0$.  The characteristic root $\omega = 1/x = \phi_1$ which must be less than one in modulus for $r_t$ to be stationary.  As shown before $\rho_t = phi_1^{\ell}$ for a stationary AR(1) model.  The condition implies that $\rho_{\ell} \longrightarrow 0$ as $\ell \longrightarrow \infty$

## 2.4.2 Identifying AR Models in Practice

The order of an AR time series is unknown: it must be specified empirically via one of two general approaches to **order determination** or order specification of AR models that determine the value of p. 
- one approach uses partial autocorrelation function
- another approach uses information criteria.

### Partial autocorrelation function (PACF)

**PACF** of a time series 
- is a function of the time series' ACF.
- is a useful tool for determining the order p of an AR model.
- is understood by considering AR models in consecutively increasing order.

$\large r_t = \phi_{0,1} + \phi_{1,1}r_{t-1} + e_{1t}$

$\large r_t = \phi_{0,2} + \phi_{1,2}r_{t-1} + \phi_{2,2}r_{t-2} + e_{2t}$

$\large r_t = \phi_{0,3} + \phi_{1,3}r_{t-1} + \phi_{2,3}r_{t-2} + \phi_{3,3}r_{t-3} + e_{3t}$

$\large r_t = \phi_{0,4} + \phi_{1,4}r_{t-1} + \phi_{2,4}r_{t-2} + \phi_{3,4}r_{t-3} + \phi_{4,4}r_{t-4} + e_{4t}$

etc ...

where j is the order of the model and 
- $\rho_{0,j}$ is the constant term
- $\phi_{i,j}$ is the coefficient of $r_{t-i}$
- $\{e_{jt}\}$ is the error term

The above AR models are in the form of a multiple linear regression and can be estimated via least squares method.

Selecting amongst them can be done vai partial F test of multiple linear regression analysis.

The estimate $\hat{\phi}_{1,1}$ is called the **lag-1 sample PACF of $\large r_t$** and is from the first equation.

The estimate $\hat{\phi}_{2,2}$ is called the **lag-2 sample PACF of $\large r_t$** and is from the second equation.

The estimate $\hat{\phi}_{3,3}$ is called the **lag-3 sample PACF of $\large r_t$** and is from the third equation.

The lag-2 PACF $\hat{\phi}_{2,2}$ is the added contribution of $\large r_{t-2}$ to r<sub>t</sub> over the AR(1) model $\large r_t = \phi_0 + \phi_1 r_{t-1} + e_{1t}$ [as provided by the AR(2) model].

The lag-3 PACF $\hat{\phi}_{3,3}$ shows the added contribution of $r_{t-3}$ [model] to r<sub>t</sub> over an AR(2) model.

Thus, an AR(p) model's lag-p sample PACF $\hat{\phi}_{3,3}$ should not be zero, but for all j > p, $\hat{\phi}_{j,j}$ should be close to zero.

[implies that partial autocorrelation is after other **earlier** lags are accounted for but not after accounting for **later** lags: $\phi_{2,3}$ would not be a lag-2 PACF of $\large r_t$ because although it accounts for correlation at lag 2 after accounting for lag effect at lag 1, it is also doing so after accounting for autocorrelation at lag-3].


This "just enough p" is the property to determine the order p: **for a stationary Gaussian AR(p) model, the sample PACF has the following properties**:
- $\hat{\phi}_{p,p}$ converges to $\phi_p$ as sample size T goes to infinity.
- $\hat{\phi}_{\ell,\ell}$ converges to zero for all $\ell$ > p.
- the asymptotic variance of $\hat{\phi}_{\ell,\ell}$ is 1/T for $\ell$ > p.
    - use the standard error associated with this variance to test PACF under the 
        - $\large H_0: \phi = 0|\ell > p$ for all p
        - $\large H_a: \phi ≠ 0|\ell > p$ means that the latest p is significant

net: the sample PACF cuts off at lag p.

example: CRSP value-weighted index Jan 1926 - Dec 2008 in table 2.1 that shows the first 12 lags of sample PACF for sample data size T = 996 that yields a standard error of the sample PACF of $sqrt{1/996}$ = 0.032.  Thus if 5% significance is used, AR(3) PACF = -0.102 has test statistic = |-0.102 / 0.032| > 1.96 for 5% and greater than 2.58 for 1% significance (qnorm(0.975) and qnorm(0.995) respectively). [pnorm(1.96)=0.975].  AR(9) PACF = -.063 has a test statistic = |-0.063 / 0.032| = 1.96 and thus is just barely not rejected at the 5% level, but is rejected at the 1% level. 



example: GNP growth rate series from Example 2.1 has figure 2.6 dotted lines to denote the approximate two standard deviation limits $\pm(2/\sqrt(176))$ 

[I compute $\pm(1.96*1/(T=\sqrt(176))) = 0.15$]. The plot suggests AR(3) model for the data because the first 3 lags of the sample PACF are large.



### Information criteria

Information criteria that determine AR order p are likelihood based.

(2.6)

**Akaike information criterion (AIC)** = $\frac{-2}{T}$ ln(likelihood) + $\frac{2}{T}$(number of parameters) 



where:
- T = sample size
- likelihood function is evaluated at the maximum likelihood estimates [$\phi$?]
- The first term in AIC in (2.16), $\large ln(\tilde{\sigma}_{\ell}^2)$, measures goodness of fit of the AR($\ell$) model to the data.
- The second term in AIC in (2.16), $\frac{2}{T}$(number of parameters) is called the **penalty function** of the information criterion lifting the AIC by a fraction of the number of parameters.
    - different penalty functions result in different information criteria.
    - in class and books, we are recommended to conjure up our own penalty functions.

for a Gaussian AR($\ell$) model, AIC reduces to **AIC($\ell$)**

$\large \text{AIC}(\ell) = ln(\tilde{\sigma}_{\ell}^2) + \frac{2*\ell}{T}$



where
- $\tilde{\sigma}_{\ell}^2$ is maximum likelihood estimate of $\sigma_a^2$ = variance of $a_t$
- T is sample size
- see [likelihoood] equation (1.18) 

(1.18)
$\large f(r_1, ..., r_T;\theta) = f(r_1;\theta) \prod_{t=2}^T \frac{1}{\sqrt{2\pi}\sigma_t}exp(\frac{-(r_t-\mu_t)^2}{2\sigma_t^2})$

The first term in AIC in (2.16), $\large ln(\tilde{\sigma}_{\ell}^2)$, measures goodness of fit of the AR($\ell$) model to the data.

**BIC** = **Schwarz-Bayesian information criterion (BIC)** for Gaussian AR($\ell$) model:



AIC penalizes by a factor of 2 times parameters.

BIC penalizes by a factor of ln(T) times parameters, $\ell$.

Thus BIC tends to select lower order AR models when the sample size is large.

### Parameter estimation

**Conditional Least Squares method** 
- starts with the (p + 1)th observation, aka "conditioning on the first p observations"
- used to estimate parameters
- in the form of multiple linear regression
- estimated with the least squares method

$$\large r_t = \phi_0 + \phi_1 r_{t-1} + ... + \phi_p r_{t-p} + a_t, \,\, t = p + 1, ..., t$$



- fitted model:

$$\large \hat{r}_t = \hat{\phi}_0 + \hat{\phi}_1 r_{t-1} + ... + \hat{\phi}_p r_{t-p} + a_t$$



- associated residual

$$\large \hat{a}_t = r_t - \hat{r}_t$$



- **residual series** is {$\hat{a}_t$} gives:

$$\large \hat{\sigma}_a^2 = \frac{\sum_{t=p+1}^T \hat{a}_t^2}{T - 2p - 1}$$



**Conditional-likelihood method** 
- same estimates of $\phi_i$ as Conditional Least Squares method
- estimate of $\sigma_a^2$ replaces the (T - 2p - 1) denominator with (T - p) or in some packages replaces (T - 2p - 1) with T.

$$\large \tilde{\sigma}_a^2 = \hat{\sigma}_t^2 \times (T - 2p - 1) / (T - p)$$

$$\large \tilde{\sigma}_a^2 = \frac{\sum_{t=p+1}^T \hat{a}_t^2}{T - p}$$

$$\large \tilde{\sigma}_a^2 = \hat{\sigma}_t^2 \times (T - 2p - 1) / (T)$$

$$\large \tilde{\sigma}_a^2 = \frac{\sum_{t=p+1}^T \hat{a}_t^2}{T}$$

### Model checking

After fitting, the model is checked for inadequacy.

Adequacy:
- adequate model has a residual series that behaves like a white noise.
    - ACF and Ljung-Box statistics (2.3) of residuals check the closeness of $\hat{a_t}$ to a white noise.
        - AR(p) model's Ljung-Box statistic Q(m) follows asymptotically a chi-squared distribution with m - g degrees of freedom where g = number of AR coefficients in the model. 
        - Adjustment in DOF [from m to m-g] is based on the [g] number of constraints added to the residuals $\hat{a_t}$ from fitting AR(p) [p=g] to an AR(0) model.
        - Most time series packages do not adjust DOF when applying Ljung-Box statistics Q(m) to a residual series. [That is, the summary statistics for residuals that the model provides will give portmanteau for m not m-g.] This is understandable when [the tested lag =] m < g [= number of coefficents].

Indadequate model must be refined.
- estimated AR coefficients that are not significantly different from zero should be removed
- the model should be extended to serial correlations that the residual ACF reveals





example: monthly value weighted simple returns AR(3) model's residual series.
Q(12) = 16.35 with m = 12 with a p-value of 0.06 based on asymptotic chi-squared distribution with m - g = 12 - 3 = lags tested - model coefficients = 9 degrees of freedom.
thus the H_0 null hypothesis of no-residual serial correlation in the first 12 lags is barely not rejected at the 5% level.  However since the lag-2 AR coefficient is not significant at the 5% level, one can refine the model, dropping the lag-2 AR coefficient which produces a model that has estimated $\phi$ that are significant at the 1% level. Now the residual series portmanteau test Ljung-Box Q(12) test of m = 12 lags = 16.83 that yields a p-value of 0.078 based on $\chi_{10}^2$ with (m=12 lags minus g=2 coefficients) = 10 degrees of freedom.  The model is **adequate in modeling dynamic linear dependence** of this data's time series **[because it leaves a white noise residual series]**.

$$\large r_t = 0.0088 + 0.114 r_{t-1} - 0.106 r_{t-3} + a_t, \,\, \hat{\sigma}_a = 0.0536$$

## 2.4.3 Goodness of fit

$$\,R^2 = 1 - \frac{SSE}{SST}$$



Common measure of **goodness of fit** of a stationary model is R squared defined as 1 minus residual sum of squared divided by total sum of squares.  For stationary AR(p) time series model with T observations {$\,r_t|t=1, ..., T$} the measure is:

$$\,R^2 = 1 - \frac{\sum_{t=p+1}^T \hat{a}_t^2}{\sum_{t=p+1}^T (r_t - \bar{r})^2}$$



where:
- $\bar{r} = \sum_{t=p+1}^T r_t / (T - p)$
- $\,0 ≤ R^2 ≤ 1$
- Typically, a larger $\,R^2$ indicates a model with a closer fit to the data but this is only true for a stationary time series.  
- For a unit-root nonstationary series, $\,R^2$ of an AR(1) fit converges to 1 when sample size increases to infinity regardless of the true underlying model of $\,r_t$.
- For a given data set, $\,R^2$ is a non decreasing function of parameter count of the model.  Adjusted $\,R^2$ overcomes this weakness and is defined as 1 minus the ratio of the variance of residuals to the variance of returns:

$$\, R_{adj}^2 = 1 - \frac{\hat{\sigma}_a^2}{\hat{\sigma}_r^2}$$



where
- $\hat{\sigma}_r^2$ is the sample variance of r<sub>t</sub>
- this measure 
    - takes account the parameter count of the fitted model.
    - no longer is bounded between 0 and 1

## 2.4.4 Forecasting

Take AR(p) model in (2.9). S

(2.9)
$$\large r_t = \phi_0 + \phi_1 r_{t-1} + ... + phi_p r_{t-p} + a_t$$



Set time index t = h and forecast to t = h + $\ell$ for $\ell$ ≥ 1 to obtain $\,r_{h+\ell}$. 
- Time index h is the **forecast origin**. 
- Positive integer $\ell$ is the **forecast horizon**.  
- Let $\hat{r}_h(\ell)$ = forecast of $\,r_{h+\ell}$
    - using minimum squared error loss function 
- In other words the forecast $\hat{r}_h(\ell)$ is chosen such that 

$$\large E[(r_{h+\ell} - \hat{r}_h(\ell))^2]|F_h ≤ min_g E[(r_{h+\ell} - g)^2|F_h]$$

where: 
- g is a function of information availble at time h (inclusive), i.e. a function of $\,F_h$
- $\hat{r}_h(\ell)$ = el-step ahead forecast of $\,r_t$ at forecast origin h
- $\,F_h$ = set of information available at forecast origin h. 



### 1-step-ahead forecast

[From the AR(p) model (2.9):]

$$\large r_t = \phi_0 + \phi_1 r_{t-1} + ... + phi_p r_{t-p} + a_t$$

[set t = h+1:]

$$\large r_{h+1} = \phi_0 + \phi_1 r_{(h+1)-1} + ... + \phi_p r_{(h+1)-p} + a_{h+1}$$

$$\large r_{h+1} = \phi_0 + \phi_1 r_{h} + ... + \phi_p r_{h+1-p} + a_{h+1}$$



[$r_{h+1}$ is a model of one-step-ahead forecast with $a_{h+1}$ innovation.  $\hat{r}_h(1)$ is the one-step-ahead forecast via Expectation where $E(\phi_0|F_h) = \phi_0$ and $E(a_{h+1}|F_h) = 0$.]  

Using minimum squared error (MSE) loss function, the point forecast of $\,r_{h+1}$ given $F_h$ is the conditional expectation

$$\large \hat{r}_h(1) = E(r_{h+1}|F_h) = \phi_0 + \sum_{i=1}^p \phi_i r_{h+1-i}$$



and the associated forecast error is

$$\large e_h(1) = r_{h+1} - \hat{r}_h(1) = a_{h+1}$$



Consequently the variance of the 1-step ahead forecast error is 

$$\large \text{Var}(e_h(1)) = \text{Var}(a_{h+1}) = \sigma_a^2$$

If $\,a_t$ is normally distributed, then a 95% 1-step ahead interval forecast of $\,r_{h+1} = \hat{r}_h(1) \pm 1.96 * \sigma_a$

For the time series in (2.4):
- $\,r_t$ is linear
- the linear model is in (2.4) with
    - $\mu$ is the mean of $\,r_t$ and 
    - {$a_t$} is a well defined distribution i.e. white noise series where $\,a_t$ denotes new information aka "innovation" or "shock" at time t of the time series
    - $\,a_{t+1}$ is also the 1-step ahead forecast error at the forecast origin t and is also called the "innovation" or "shock" of the series at time t+1.

(2.4)
$$\large r_t = \mu + \sum_{i=0}^\infty \psi_i a_{t-i}$$

Estimated parameters are used to compute point and interval forecasts resulting in a **conditional forecast** because such a foecast does not take into consideration the uncertainty of the parameter estimates.  One could take this into consideratin but it is much more complicated. A natural way to consider parameter and model uncertainty in forecasting is Bayesian forecasting with Markov chain Monte Carlo (MCMC) methods discussed in chapter 12.  Here, assume simply that the model is given. When the sample size for estimation is large, the conditional forecast is close to the unconditional forecast. 
    
### 2-step-ahead forecast

$\large r_{h+2}$ forecast from the forecast origin h is the **2-step forecast**.  Using AR(p) again [to form a model]:

$$\large r_{h+2} = \phi_0 + \phi_1 r_{h+1} + ... + \phi_p r_{h+2-p} + a_{h+2}$$



Taking conditional expecation:

$$\large \hat{r}(2) = E(r_{h+2}|F_h) = \phi_0 + \phi_1 \hat{r}_{h+1} + ... + \phi_p \hat{r}_{h+2-p}$$



and associated forecast error

$$\large e_h(2) = r_{h+2} - \hat{r}_h(2) = \phi_1 [r_{h+1} - \hat{r}_h(1)] + a_{h+2} = a_{h+2} + \phi_1 a_{h+1}$$



The variance of the forecast error 

$$\large Var[e_h(2)] = (1 + \phi^2) \sigma_a^2$$

[I assume this is so because $\sigma_{a_{h+2}}^2 = \sigma_{a_{h+1}}^2$]

Interval forecasts of $\,r_{h+2}$ are computed in same way as $\,r_{h+1}$:  "If $a_t$ is normally distributed, then a 95% 1-step ahead interval forecast of $r_{h+1} = \hat{r}_h(1) \pm 1.96 * \sigma_a$"

[Notice that with the above formulas for Var(e_h(i))] that as the forecast horizon increases, the uncertainty in the forecast also increases: at the time index h, the forecast for $r_{h+2}$ is more uncertain than the forecast for $r_{h+1}$:

$\large \text{Var}[e_h(2)] = (1 + \phi^2) \sigma_a^2 ≥ \text{Var}(e_h(1)) = \text{Var}(a_{h+1}) = \sigma_a^2$


### Multistep-ahead forecast

General [expression for the forecast model:]

$$\large r+{h+\ell} = \phi_0 + \phi_1 r_{h+\ell-1} + ... + \phi_p r_{h+\ell-p} + a_{h+\ell}$$



The **$\ell$-step ahead forecast**
- based on minimum squared error loss function
- is the **conditional expectation** of $\,r_{h+\ell}$ given $F_h$  
- obtained by

$$\large \hat{r}_h(\ell) = \phi_0 + \sum_{i=1}^p \phi_i \hat{r}_h(\ell - i)$$



- where $\hat{r}_h(i) = r_{h+i} \text{ if } i < 0$ [if i is in the past not future]

The forecast is then computed recursively using forecasts for $\hat{r}_h(i)$ for i = 1, ..., $\ell-1$.

The **$\ell$-step ahead forecast error** is 

$$\large e_h(\ell) = r_{h+\ell} - \hat{r}_h(\ell)$$

**Mean reversion** For a stationary AR(p) model, $\hat{r}_h(\ell)$ converges to $\,E(r_t)$ as $\ell$ --> $\infty$, meaning that for such a series [stationary?] the long-term porint forecast $\hat{r}_h(\ell)$ approaches its unconditional mean $\,E(r_t)$.  This property is known as **mean reversion**.

[so mean reversion occurs as $\ell$ goes to infinity; how far do we need to forecast, how big an $\ell$ do we need, to see the effect of mean reversion? ... next paragraph...]

For an AR(1) model [p=1] the speed of mean reversion [of the long term forecast approach to its unconditional mean] is measured by the **half-life** defined as 

$$\large \ell = \frac{ln(0.5)}{ln(|\phi_1|)}$$

The **variance of the forecast error** approaches the unconditional variance of $\,r_t$. 

Note that for an AR(1) model in (2.8), 

$$\large r_t = \phi_0 + \phi_1 r_{t-1} + a_t$$

let 
- $x_t = r_t - E(r_t)$ be the mean-mean adjusted series.

then its easy to see that the $\ell$-step ahead forecast of $x_{h+\ell}$ at the forecast origin h is ...

$\large \hat{x}_h(\ell) = \phi_1^{\ell}$

[... because you are recursively substituting in the same linear for $\,r_{t-1}$]

[here is deriving that formula for the forecast horizon $\ell$ at which half life occurs.]

The half life is the forecast horizon such that

$\large \hat{x}_h(\ell) = \frac{1}{2} x_h$

because 

$\large \phi_1^{el} = \frac{1}{2}$

Thus 

$\large \ell = \frac{ln(0.5)}{ln|\phi_1|)}$

[It is half of the horizon value of x because x is mean adjusted r and the forecast is thus approaching its mean which for x is zero.]
